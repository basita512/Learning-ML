## Using OO method (Object Oriented API)

### Which one should you use? (pyplot or matplotlib OO method)

1. When plotting something quickly use pyplot mehtod
2. When plotting something more advanced use the OO mehtod

```python
over_50 = heart[heart['age'] > 50]
over_50.head(10)
```

```python
# PyPlot method
over_50.plot(kind='scatter', x='age', y='chol', c='target')
# here c=color
```

```python
## OO method mix with pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax)
ax.set_xlim([40, 100])

# set_xlim sets the limit of x axis from 40 to 100
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

## adding data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])

# cutomizing the plot 
ax.set(title='Cholestrol level of patients',
      xlabel='Age',
      ylabel='Cholestrol')

# Adding legend (Legend means the key which we used to have for graph)
ax.legend(*scatter.legend_elements(), title='Target')

# Adding a horizontal line
ax.axhline(y=over_50['chol'].mean(), linestyle='--')
```

```python
over_50.head()
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)

# Adda daata to ax0
scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])

#customize the ax0
ax0.set(title='Heart Disease cholerstrol levels', xlabel='Age', ylabel='Cholestrol')

#adding legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

#Adding a meanline
ax0.axhline(y=over_50['chol'].mean(), linestyle='--')


# for ax1
scatter2 = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'])
ax1.set(title='Heart disease thalanch levels', xlabel='Age', ylabel='Max Heart Rate')
ax1.legend(*scatter2.legend_elements(), title='Target')
ax1.axhline(y=over_50['thalach'].mean())

fig.suptitle('Heart Disease Analysis', fontsize='20', fontweight='bold')
```
