# Matplotlib example Workflow

## 2. Setup plot

### fig, ax = plt.subplots(figsize=(10, 5))

plt.subplots()  creates a figure and a set of axes (a plotting area).

fig  the entire canvas or figure, like the blank paper.

ax  the plot area inside the figure where the graph actually appears.

figsize=(10, 5)  sets the width and height of the figure in inches (10 inches wide, 5 inches tall).

## 3. Plot data

### ax.plot(x, y)

ax.plot(...)  tells matplotlib to draw a line graph on the axes.

It connects the (x, y) points with lines.

This is equivalent to saying: Hey ax, draw a line that connects these data points.

```python
# 0. Import matplotlib and get redy for jupyter forplotting
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [51, 42, 32, 46]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 5))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title='Simple Plot',
      xlabel='axis-X',
      ylabel='axis-Y')

# 5. Save and show (save the whole figure)
fig.savefig('../Matplotlib/images/sample-plot.png')
```

```python
x = [15, 16, 13, 12]
y = [2, 4, 6, 8]

#Setup plot
fig, ax = plt.subplots(figsize=(10,5))

#plot data
ax.plot(x, y)

#customize
ax.set(title='Sample-2',
      xlabel='I am X-axis',
      ylabel='I am Y-axis')

#save
fig.savefig('../Matplotlib/images/sample-2.png')
```

# Making figures with Numpy array

```python
import numpy as np
```

```python
# creating some random data
# np.linspace(start, stop, num=50)
# Its a method from the NumPy library used to generate evenly spaced numbers over a specific range.
x = np.linspace(0, 10, 100)
x[:20]
```

## Line plot

```python
# plot data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

## Scatter plot

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

## Bar plot

```python
# Making from a dicionary
prices = {'Car':20,
         'toy': 50,
         'box':10}
fig, ax = plt.subplots()
ax.bar(prices.keys(), prices.values())
ax.set(title='Prices List',
      xlabel='Things',
      ylabel='Prices')
fig.savefig('../Matplotlib/images/prices.png')
```

```python
# Horizontal Bar graph
fid, ax = plt.subplots()
ax.barh(list(prices.keys()), list(prices.values()))
```

## Histograms

```python
# make data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

#Plot to each different subplot
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(prices.keys(), prices.values())
ax4.hist(np.random.randn(5000))
```
